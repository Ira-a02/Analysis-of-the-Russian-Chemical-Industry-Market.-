```python
import random
def data_discount_settings(row):
    category = row['Категория']
    price = row['Цена']
    if not pd.isna(row['Скидочная Цена']):
        return row['Скидочная Цена']
    if category == 'Газы и газовые смеси':
        discounts = range(3,11)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Промышленные технологии':
        discounts = range(5,16)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Косметика':
        discounts = range(5,26)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Герметики':
        discounts = range(10,21)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Клеи и строительные смеси':
        discounts = range(10,31)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Сварочные материалы и оборудование':
        discounts = range(5,21)
        return round(price - price * (random.choice(discounts)/100))
    return row['Скидочная Цена']
data['Скидочная Цена'] = data.apply(data_discount_settings, axis = 1, result_type = 'expand')
```
