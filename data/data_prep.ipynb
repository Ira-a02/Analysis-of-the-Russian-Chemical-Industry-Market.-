```python
import random
def data_discount_settings(row):
    category = row['Категория']
    price = row['Цена']
    if not pd.isna(row['Скидочная Цена']):
        return row['Скидочная Цена']
    if category == 'Газы и газовые смеси':
        discounts = range(3,11)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Промышленные технологии':
        discounts = range(5,16)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Косметика':
        discounts = range(5,26)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Герметики':
        discounts = range(10,21)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Клеи и строительные смеси':
        discounts = range(10,31)
        return round(price - price * (random.choice(discounts)/100))
    if category == 'Сварочные материалы и оборудование':
        discounts = range(5,21)
        return round(price - price * (random.choice(discounts)/100))
    return row['Скидочная Цена']
data['Скидочная Цена'] = data.apply(data_discount_settings, axis = 1, result_type = 'expand')
```

```python
# Генерация истории цен на основе Индексов цен с января 2020 по октябрь 2025
# Индексы взяты от раздела "Производство химических веществ и химических продуктов" с Росстата 

monthly_indexes = {
    # 2024-2025 (октябрь 2024 → октябрь 2025)
    '2025-01': 103.01, '2025-02': 101.60, '2025-03': 102.41, '2025-04': 99.23,
    '2025-05': 97.14, '2025-06': 95.80, '2025-07': 101.33, '2025-08': 96.05,
    '2025-09': 101.42, '2025-10': 101.04,
    
    # 2023-2024 
    '2024-01': 98.64, '2024-02': 98.61, '2024-03': 102.22, '2024-04': 101.75,
    '2024-05': 101.00, '2024-06': 99.54, '2024-07': 99.94, '2024-08': 100.65,
    '2024-09': 101.13, '2024-10': 100.62, '2024-11': 100.18, '2024-12': 100.76,
    
    # 2022-2023
    '2023-01': 100.56, '2023-02': 100.48, '2023-03': 99.73, '2023-04': 100.94,
    '2023-05': 101.78, '2023-06': 97.88, '2023-07': 100.09, '2023-08': 102.32,
    '2023-09': 103.45, '2023-10': 101.96, '2023-11': 100.74, '2023-12': 97.93,
    
    # 2021-2022
    '2022-01': 104.83, '2022-02': 100.84, '2022-03': 103.97, '2022-04': 105.57,
    '2022-05': 102.56, '2022-06': 95.42, '2022-07': 92.71, '2022-08': 94.53,
    '2022-09': 93.91, '2022-10': 97.07, '2022-11': 97.90, '2022-12': 98.68,
    
    # 2020-2021
    '2021-01': 103.47, '2021-02': 104.48, '2021-03': 105.45, '2021-04': 107.53,
    '2021-05': 106.88, '2021-06': 102.09, '2021-07': 101.06, '2021-08': 101.62,
    '2021-09': 100.86, '2021-10': 101.74, '2021-11': 104.25, '2021-12': 102.90,
}

def get_previous_month(month_str):
    """Получить предыдущий месяц в формате YYYY-MM"""
    year, month = map(int, month_str.split('-'))
    if month == 1:
        return f"{year-1}-12"
    else:
        return f"{year}-{month-1:02d}"
def calculate_chain_prices(current_price, start_month='2025-10', months_back=60):
    """
    Рассчитывает цены назад по цепочке месяцев
    """
    prices = {}
    current_price_float = current_price
    
    # Текущий месяц
    prices[start_month] = current_price_float
    current = start_month
    
    # Идем назад по месяцам
    for i in range(months_back):
        prev_month = get_previous_month(current)
        
        if prev_month in monthly_indexes:
            # Цена предыдущего месяца = текущая цена / индекс
            index = monthly_indexes[current] / 100  # индекс для перехода current → prev
            current_price_float = current_price_float / index
            prices[prev_month] = current_price_float
            current = prev_month
        else:
            break
    
    return prices
current_price = 1000  # Цена в октябре 2025
historical_prices = calculate_chain_prices(current_price)
```

```python

def create_quarterly_features(historical_prices):
    """
    Создает квартальные фичи на основе месячных данных
    Берет последний месяц каждого квартала
    """
    quarterly_features = {}
    
    # Отбираем конечные месяцы кварталов
    quarter_months = {
        'Q1': '03',  # март
        'Q2': '06',  # июнь
        'Q3': '09',  # сентябрь
        'Q4': '12'   # декабрь
    }
    
    for month, price in historical_prices.items():
        year, month_num = month.split('-')
        
        # Проверяем, является ли месяц конечным в квартале
        for quarter, end_month in quarter_months.items():
            if month_num == end_month:
                feature_name = f'Цена_{quarter}_{year}'
                quarterly_features[feature_name] = round(price, 2)
    
    return quarterly_features

# Создаем квартальные фичи
quarterly_features = create_quarterly_features(historical_prices)
print(quarterly_features)
```

```python
# Определяем порядок кварталов (от новых к старым)
quarters_order = [
        'Цена_Q3_2025', 'Цена_Q2_2025', 'Цена_Q1_2025',
        'Цена_Q4_2024', 'Цена_Q3_2024', 'Цена_Q2_2024', 'Цена_Q1_2024', 
        'Цена_Q4_2023', 'Цена_Q3_2023', 'Цена_Q2_2023', 'Цена_Q1_2023',
        'Цена_Q4_2022', 'Цена_Q3_2022', 'Цена_Q2_2022', 'Цена_Q1_2022',
        'Цена_Q4_2021', 'Цена_Q3_2021', 'Цена_Q2_2021', 'Цена_Q1_2021'
    ]

# Определяем уровень волатильности для каждой категории

'''
Обоснование уровней волатильности:

Высокая волатильность (6-8%):
Газы и газовые смеси (8%) - сильно зависят от цен на энергоносители, сырьевые рынки, логистики
Клеи и строительные смеси (7%) - тесно связаны со строительным циклом, сезонностью спроса
Средняя волатильность (4-6%):

Герметики (6%) - зависимость от строительного рынка, но более стабильный спрос
Промышленные технологии (5%) - B2B сегмент с долгосрочными контрактами
Низкая волатильность (3-4%):

Косметика (3%) - потребительские товары с устойчивым спросом, брендовая лояльность
Сварочные материалы (4%) - промышленное применение, но специализированный спрос
'''

category_noise = {
    'Газы и газовые смеси': (0.95, 1.10),       # -5% до +10% - высокая волатильность
    'Промышленные технологии': (0.98, 1.07),    # -2% до +7% - средняя волатильность  
    'Косметика': (0.99, 1.05),                  # -1% до +5% - низкая волатильность
    'Герметики': (0.97, 1.08),                  # -3% до +8% - средняя волатильность
    'Клеи и строительные смеси': (0.96, 1.09),  # -4% до +9% - высокая волатильность
    'Сварочные материалы и оборудование': (0.98, 1.06),  # -2% до +6% - низкая волатильность
}


def calculate_quarterly_tuple(row):
    """
    Функция для apply, возвращает кортеж квартальных цен
    """
    current_price = row['Цена']
    current_category = row['Категория']
    current_noise = category_noise.get(current_category, (0.97, 1.05))
    
    historical_prices = calculate_chain_prices(current_price)
    quarterly_features = create_quarterly_features(historical_prices)
    
    quarterly_features_noise = {}
    for quarter, price in quarterly_features.items():
        noise = np.random.uniform(current_noise[0], current_noise[1])
        price_with_noise = price * noise
        quarterly_features_noise[quarter] = round(price_with_noise)
    
    # Создаем кортеж значений в нужном порядке
    values_tuple = tuple(
        quarterly_features_noise.get(quarter, None)
        for quarter in quarters_order)
    
    return values_tuple

quarters_prices = data.apply(calculate_quarterly_tuple, axis = 1, result_type = 'expand')
quarters_prices_pd = pd.DataFrame(quarters_prices)
quarters_prices_pd = quarters_prices_pd.rename(columns = {i: col for i,col in enumerate(quarters_order)})
```

```python
companies_modal_category = data.groupby(['Компания'])['Категория'].agg(lambda x: x.value_counts().index[0]).to_frame()
companies_modal_category = companies_modal_category.rename(columns = {'Категория':'Модальная категория'})

companies_modal_category_size = data.groupby(['Компания'])['Категория'].agg(lambda x: x.value_counts().iloc[0]).to_frame()
companies_modal_category_size = companies_modal_category_size.rename(columns = {'Категория':'Кол-во товаров модальной категории'})

companies_median_discount = data.groupby('Компания')['Процент скидки'].agg('median').to_frame()
companies_median_discount = companies_median_discount.rename(columns = {'Процент скидки':'Медианный % скидки'})

data_companies = pd.concat([companies_modal_category,companies_modal_category_size,companies_median_discount], axis = 1)
```

```python
companies_categories = data.pivot_table(
    index = 'Компания',
    columns = 'Категория',
    values = 'Название товара',
    aggfunc = 'count',
    fill_value = 0
).reset_index()

companies_categories = companies_categories.rename(columns = {
    'Газы и газовые смеси':'Кол-во товара: Газы и газовые смеси',
    'Герметики':'Кол-во товара: Герметики',
    'Клеи и строительные смеси':'Кол-во товара: Клеи и строительные смеси',
    'Косметика':'Кол-во товара: Косметика',
    'Промышленные технологии':'Кол-во товара: Промышленные технологии',
    'Сварочные материалы и оборудование':'Кол-во товара: Сварочные материалы и оборудование',
})
companies_categories = companies_categories.set_index('Компания', drop = True)
companies_categories
```

```python
companies_discounts = data.pivot_table(
    index = 'Компания',
    columns = 'Категория',
    values = 'Процент скидки',
    aggfunc = 'median',
    fill_value = np.nan
).reset_index()

companies_discounts = companies_discounts.rename(columns = {
    'Газы и газовые смеси':'Медианный % скидки: Газы и газовые смеси',
    'Герметики':'Медианный % скидки: Герметики',
    'Клеи и строительные смеси':'Медианный % скидки: Клеи и строительные смеси',
    'Косметика':'Медианный % скидки: Косметика',
    'Промышленные технологии':'Медианный % скидки: Промышленные технологии',
    'Сварочные материалы и оборудование':'Медианный % скидки: Сварочные материалы и оборудование',
})
companies_discounts = companies_discounts.set_index('Компания', drop = True)
companies_discounts
```

```python
data_companies = pd.concat([data_companies,companies_categories, companies_discounts], axis = 1)
data_companies.to_csv(FOLDER_PATH + 'data_companies.csv')
data_companies.to_excel(FOLDER_PATH + 'data_companies.xlsx')
data_companies
```
