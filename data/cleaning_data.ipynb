## Altaystroy

```python
altaystroy = pd.read_excel(FOLDER_PATH + 'АЛТАЙСТРОЙДЕТАЛЬ.xlsx')
```

```python
def altaystroy_volume_prep(item):
    if pd.isna(item):
        return None
    if 'кг' in item:
        return 'кг'
    if 'м3' in item:
        return 'м3'
    if 'баллон' in item:
        return 'баллон'
        
altaystroy['Цена'] = altaystroy['Цена'].str.extract(r'([\d.]+)').astype('float64')
altaystroy = altaystroy.rename(columns = {'Объем':'Единицы Измерения'})

altaystroy['Единицы Измерения'] = altaystroy['Единицы Измерения'].apply(altaystroy_volume_prep)

altaystroy.loc[altaystroy['Единицы Измерения'].isna(),'Единицы Измерения'] = np.nan
altaystroy.loc[altaystroy['Цена'].isna(),'Цена'] = np.nan

altaystroy['Название товара'] = altaystroy['Название товара'].apply(lambda x: x.strip())
altaystroy['Категория'] = altaystroy['Категория'].apply(lambda x: x.strip())
altaystroy['Компания'] = 'АЛТАЙСТРОЙДЕТАЛЬ'
altaystroy['Объем'] = np.nan
altaystroy['Скидочная Цена'] = np.nan
```

## Kesto

```python
kesto = pd.read_excel(FOLDER_PATH + 'КЕСТО_cleaned.xlsx')
```

```python
import re

def kesto_str_to_lst(item):
    if pd.isna(item):
        return None
    lst = item[1:-2]
    lst = re.split(r'[,;]', lst)
    lst = [x for x in lst if 'nan' not in x]
    if len(lst) == 0:
        return None
    return lst

def kesto_units_of_measurement(item):
    if item == None: 
        return None
    if 'мл' in item[0]:
        return 'мл'
    if 'л' in item[0]:
        return 'литр'
    if 'кг' in item[0]:
        return 'кг'
    if 'м.кв' in item[0]:
        return 'м2'
def kesto_no_measurement(item):
    if item == None: 
        return None
    items_cleaned = list(map(lambda x: re.search(r'([\d]+)', x).groups() if re.search(r'([\d]+)', x) != None else None, item))
    items_cleaned = list(map(lambda x: float(x[0]) if x!= None else None,items_cleaned))
    return items_cleaned
    
kesto['Объем'] = kesto['Объем'].apply(kesto_str_to_lst)    
kesto['Единицы Измерения'] = kesto['Объем'].apply(kesto_units_of_measurement)
kesto['Объем'] = kesto['Объем'].apply(kesto_no_measurement)
kesto.loc[kesto['Единицы Измерения'] == 'мл','Объем'] = kesto.loc[kesto['Единицы Измерения'] == 'мл','Объем'].apply(lambda x: int(x[0])/1000)
kesto.loc[kesto['Единицы Измерения'] == 'мл','Единицы Измерения'] = 'литр'

kesto.loc[kesto['Объем'].isna(),'Объем'] = np.nan
kesto.loc[kesto['Единицы Измерения'].isna(),'Единицы Измерения'] = np.nan

kesto['Название товара'] = kesto['Название товара'].apply(lambda x: x.strip())
kesto['Категория'] = kesto['Категория'].apply(lambda x: x.strip())
kesto['Компания'] = 'КЕСТО'
kesto['Цена'] = np.nan
kesto['Скидочная Цена'] = np.nan
```

## Crio Service

```python
crio_service = pd.read_excel(FOLDER_PATH + 'КРИО_СЕРВИС_cleaned.xlsx')
```

```python
import re
crio_service['Цена'] = crio_service['Цена'].astype('float64')
def crio_service_volume_prep(item):
    if pd.isna(item):
        return None
    data = re.findall(r'(\d+)', item)
    for i in range(len(data)):
        data[i] = float(data[i])
    return data
crio_service['Объем'] = crio_service['Объем'].apply(crio_service_volume_prep)
crio_service['Единицы Измерения'] = 'литр'

crio_service.loc[crio_service['Объем'].isna(),'Объем'] = np.nan

crio_service['Название товара'] = crio_service['Название товара'].apply(lambda x: x.strip())
crio_service['Компания'] = 'КРИО_СЕРВИС'
crio_service['Категория'] = np.nan
crio_service['Скидочная Цена'] = np.nan
```

## Crimean Rose

```python
crimean_rose = pd.read_excel(FOLDER_PATH + 'КРЫМСКАЯ-РОЗА.xlsx')
```

```python
import re

def crimean_rose_price_prep(row):
    if pd.isna(row['Цена']):
        return None
    data = re.findall(r'([\d.]+)', row['Цена'])
    for i in range(len(data)):
        data[i] = float(data[i])
        
    if len(data)>1:
        return max(data), min(data)
    else:
        return data[0], None
def crimean_rose_volume_prep(row):
    if pd.isna(row['Объем']):
            return None, None
    data = re.findall(r'(\d+)', row['Объем'])
    data[0] = float(data[0])/ 1000
    return data, 'литр'  

crimean_rose['Название товара'] = crimean_rose['Название товара'].apply(lambda x: x.strip())
crimean_rose['Категория'] = crimean_rose['Категория'].apply(lambda x: x.strip())

crimean_rose['Скидочная Цена'] = None
crimean_rose[['Цена','Скидочная Цена']] = crimean_rose[['Цена','Скидочная Цена']].apply(crimean_rose_price_prep,
                                                                                        axis = 1,
                                                                                        result_type = 'expand')
crimean_rose['Единицы Измерения'] = None
crimean_rose[['Объем','Единицы Измерения']] = crimean_rose[['Объем','Единицы Измерения']].apply(crimean_rose_volume_prep,
                                                                                                axis = 1,
                                                                                                result_type = 'expand')

crimean_rose.loc[crimean_rose['Объем'].isna(),'Объем'] = np.nan
crimean_rose.loc[crimean_rose['Скидочная Цена'].isna(),'Скидочная Цена'] = np.nan
crimean_rose.loc[crimean_rose['Единицы Измерения'].isna(),'Единицы Измерения'] = np.nan

crimean_rose['Компания'] = 'КРЫМСКАЯ-РОЗА'
```

## Beliff Gas

```python
beliff = pd.read_excel(FOLDER_PATH + 'АО_АПК_БЕЛИФФ_ГАЗ_cleaned.xlsx')
```

```python
beliff = beliff.drop(columns = ['Вид']).reset_index(drop = True)
```

```python
import re
def beliff_price_prep(item):
    if pd.isna(item):
        return np.nan
    price = re.findall(r'([\d]+)',str(item))
    if len(price) == 0:
        return np.nan
    return ''.join(price)
beliff['Цена'] = beliff['Цена'].apply(beliff_price_prep)
beliff['Цена'] = beliff['Цена'].astype('float64')

def beliff_volume_prep(row):
    item = row['Объем']
    measure = row['Единицы Измерения']
    name = row['Название']
    category = name.split(',')[0].strip()

    if 'Поставка газов' in name:
        if 'Газы' in name:
            category = name.split(',')[1].strip()
        elif 'Жидкие газы' in name:
            category = name.split(',')[1].strip()
        elif 'Смеси' in name:
            category = name.split(',')[2].strip()
            
    elif 'Баллоны и баллонное оборудование' in name:
        category = name.split(',')[2].strip()
        
    elif 'Криогенное оборудование' in name:
        category = name.split(',')[1].strip()
        
    elif 'Сварочные материалы' in name:
        category = name.split(',')[1].strip()    
    if pd.isna(item):
        return np.nan, np.nan, category
        
    if 'кг' in item:
        volume_kg = re.findall(r'([\d.]+)',item)
        return (volume_kg[0],'кг', category) if volume_kg else (np.nan, 'кг', category)
        
    if (re.search(r'(\bл\b)',item)) or ('литров' in item):
        volume_liter = re.findall(r'([\d.]+)',str(item))
        return (volume_liter[0], 'литр', category) if volume_liter else (np.nan, 'литр', category)
    if 'шт' in item:
        volume_st = re.findall(r'([\d.]+)', str(item))
        return (volume_st[0], 'кг', category) if volume_st else (np.nan, 'кг', category)
        
    if bool(re.match(r'(^\d+\.\d+$)',str(item).strip())):
        return str(item).strip(), 'кг', category
    return np.nan, np.nan, category
    
def beliff_category_prep(row):
    anem = row['Название']
    return item.split(',')[0]
beliff['Единицы Измерения'] = None
beliff['Категория'] = None
beliff['Скидочная Цена'] = None
beliff[['Объем','Единицы Измерения', 'Категория']] = beliff.apply(beliff_volume_prep, axis = 1, result_type = 'expand')
beliff['Компания'] = 'АО_АПК_БЕЛИФФ_ГАЗ'

```
